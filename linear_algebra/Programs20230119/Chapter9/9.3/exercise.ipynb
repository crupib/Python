```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
p = array((0, 0))
q = array((2, 0))
t = linspace(0, 5, 1000)
x1 = 1 + sin(sqrt(2)*t)/sqrt(2)
x2 = (sqrt(2)*cos(sqrt(2)*t) + sin(sqrt(2)*t)) / (2*sqrt(2))
v1 = cos(sqrt(2*t))
v2 = (-sqrt(2)*sin(sqrt(2)*t) + cos(sqrt(2)*t)) / 2
```

```python
plt.axis('scaled'), plt.xlim(-0.5,2.5), plt.ylim(-1,1.5)
plt.scatter(0, 0)
plt.scatter(2, 0)
plt.scatter(x1[0], x2[0])
plt.plot([0, x1[0]], [0, x2[0]])
plt.plot([2, x1[0]], [0, x2[0]])
plt.quiver(x1[0],  x2[0], v1[0], v2[0], units='xy', scale=1)
plt.text(0, 0, 'p', fontsize = 20, verticalalignment ='top', horizontalalignment ='center')
plt.text(2, 0, 'q', fontsize = 20, verticalalignment ='top', horizontalalignment ='center')
plt.text(x1[0], x2[0], 'x(0)', fontsize = 20, verticalalignment ='bottom', horizontalalignment ='right')
plt.text(x1[0]+v1[0], x2[0]+v2[0], 'v(0)', fontsize = 20)
plt.plot(x1, x2)
```
